```python
db_file = 'food_wastage.db'  # Choose a name for your database file
table_name = 'providers'
```

```python
try:
    # Connect to SQLite
    conn = _sqlite3.connect(db_file)
    cursor = conn.cursor()
    print(f"Connected to {db_file}")

    # Create table (if it doesn't exist) - Define your columns and their data types
    cursor.execute(f"""
        CREATE TABLE IF NOT EXISTS {table_name} (
            Provider_ID INTEGER PRIMARY KEY,
            Name TEXT,
            Type TEXT,
            Address TEXT,
            City TEXT,
            Contact TEXT
        )
    """)
    conn.commit()
    print(f"Table '{table_name}' created (if needed).")

    # Transfer data from DataFrame to SQLite Table using Placeholders
    cols = ', '.join(providers.columns)  # Get column names for the INSERT statement
    placeholders = ', '.join(['?'] * len(providers.columns))
    sql = f"INSERT INTO {table_name} ({cols}) VALUES ({placeholders})"

    for index, row in providers.iterrows():
        values = tuple(row)
        cursor.execute(sql, values)

    conn.commit()
    print(f"Data from DataFrame '{providers.__class__.__name__}' inserted into '{table_name}'.")

except _sqlite3.Error as e:
    print(f"Database error: {e}")
    if conn:
        conn.rollback()

finally:
    if conn:
        conn.close()
        print("Connection closed.")
```

```python

Query_1 = """
    select count(p.Provider_ID) as Providers_Count ,count(r.Receiver_ID) as Receivers_Count ,p.City FROM providers p
JOIN receivers r on p.City=r.City
group by p.City
order by p.City ASC;
    """
df = pd.read_sql_query(Query_1, con=conn)
print(df)
```

```python
Query_2 = """
SELECT Provider_Type,Quantity FROM food_listings
GROUP by Provider_Type
ORDER by Quantity DESC ;
"""
df=pd.read_sql_query(Query_2,con=conn)
print(df)
```

```python
Query_3 = """  
SELECT Name,City,Contact FROM providers
GROUP by City;
"""
df=pd.read_sql_query(Query_3,con=conn)
print(df)
```

```python
Query_4="""  
SELECT count(c.Food_ID) as TotalFoodClaimed,r.Name,r.Receiver_ID from claims_data c
JOIN receivers r on r.Receiver_ID=c.Receiver_ID
WHERE c.Status='Completed'
GROUP by r.Name,r.Receiver_ID
ORDER by TotalFoodClaimed DESC;
"""
df=pd.read_sql_query(Query_4,con=conn)
print(df)
```

```python
Query_5="""  
SELECT f.Quantity,f.Provider_ID,p.Name FROM food_listings f
JOIN providers p on f.Provider_ID=p.Provider_ID
GROUP by p.Provider_ID,p.Name
ORDER by f.Quantity DESC;
"""
df=pd.read_sql_query(Query_5,con=conn)
print(df)
```

```python
Query_6='''  
SELECT Location,count(Food_ID) as CountOfFoodListings FROM food_listings
GROUP by Location
ORDER by CountOfFoodListings DESC;
'''
df=pd.read_sql_query(Query_6,con=conn)
print(df)
```

```python
Query_7='''  
SELECT count(Food_ID) as AvaialableQuantity,Food_Type from food_listings
GROUP by Food_Type
ORDER by AvaialableQuantity DESC;
'''
df=pd.read_sql_query(Query_7,con=conn)
print(df)
```

```python
Query_8=''' 
SELECT count(c.Claim_ID) as Count_Of_Claims,f.Food_Name
FROM claims_data c 
JOIN food_listings f on c.Food_ID=f.Food_ID 
GROUP by f.Food_Name
order by Count_Of_Claims ASC
'''
df=pd.read_sql_query(Query_8,con=conn)
print(df)
```

```python
Query_9=''' SELECT p.Name, count(c.Claim_ID) as SuccesfullClaims
FROM claims_data c
JOIN food_listings f on c.Food_ID=f.Food_ID
JOIN providers p on p.Provider_ID=f.Provider_ID
WHERE c.Status='Completed'
GROUP by p.Name 
ORDER by SuccesfullClaims DESC'''

df=pd.read_sql_query(Query_9,con=conn)
print(df)
```

```python
Query_10=''' SELECT c.Receiver_ID, avg(f.Quantity) as AvgQantityOFFood FROM claims_data c
JOIN food_listings f on c.Food_ID=f.Food_ID
GROUP by Receiver_ID'''
df=pd.read_sql_query(Query_10,con=conn)
print(df)
```

```python
Query_11=''' SELECT f.Meal_Type, count(c.Claim_ID) as NumberOfClaims FROM food_listings f
JOIN claims_data c on c.Food_ID=f.Food_ID
GROUP by f.Meal_Type
ORDER by NumberOfClaims DESC '''
df=pd.read_sql_query(Query_11,con=conn)
print(df)
```

```python
Query_12=''' SELECT p.Provider_ID,sum(f.Quantity) as TotalQuantityDonated FROM providers p
JOIN food_listings f on p.Provider_ID=f.Provider_ID
GROUP by p.Provider_ID
ORDER by TotalQuantityDonated DESC'''
df=pd.read_sql_query(Query_12,con=conn)
print(df)
```
